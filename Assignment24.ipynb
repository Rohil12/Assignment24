```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
```

```python
#Q.2
plt.scatter(x,y)
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my title")
```

```python
#Q.3
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x,y1,x,y2,x,y3,x,y4)

```

```python
#Q.4
# A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. A bar chart describes the comparisons between the discrete categories. One of the axis of the plot represents the specific categories being compared, while the other axis represents the measured values corresponding to those categories.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
```

```python
#Q.5
#A box plot (or box-and-whisker plot) is a graphical representation of the distribution of a dataset. It displays the datasets minimum, first quartile (Q1), median, third quartile (Q3), and maximum values. Box plots are particularly useful for identifying outliers and understanding the spread and skewness of the data.
import matplotlib.pyplot as plt
import numpy as np

# Sample data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot
plt.boxplot([box1, box2])

# Add labels and title
plt.xlabel('Data Sets')
plt.ylabel('Values')
plt.title('Box Plot of Two Data Sets')

# Show the plot
plt.show()
```
